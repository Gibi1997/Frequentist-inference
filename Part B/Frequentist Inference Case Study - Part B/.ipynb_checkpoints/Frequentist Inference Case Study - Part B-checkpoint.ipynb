Calculate the critical value and the relevant 95% confidence interval for the mean, and comment on whether the administrator should be concerned.

__A:__

Since we only need to know whether the  actual average charge has fallen below 12,000, we only want to find out the one-sided confidence interval; the lower confidence bond.

```python
# Given values
mean = 13270.422265141257
std = 12110.011236693994
n = 1338

# Standard error
SE = std / np.sqrt(n)

# One-sided 95% t critical value (df = n-1)
t_crit = t.ppf(0.95, df=n-1)

# Lower bound of the one-sided 95% CI
lower_bound = mean - t_crit * SE

print("Standard Error:", SE)
print("t critical (one-sided 95%):", t_crit)
print("Lower bound of 95% CI:", lower_bound)
```

The administrator should not be concerned as lower bound of 95% CI is 12725.48718381623; which means population mean will not be less than this value.

__Q4:__ State the null and alternative hypothesis here. Use the _t_-test for the difference between means, where the pooled standard deviation of the two groups is given by:
\begin{equation}
s_p = \sqrt{\frac{(n_0 - 1)s^2_0 + (n_1 - 1)s^2_1}{n_0 + n_1 - 2}}
\end{equation}

and the *t*-test statistic is then given by:

\begin{equation}
t = \frac{\bar{x}_0 - \bar{x}_1}{s_p \sqrt{1/n_0 + 1/n_1}}.
\end{equation}

(If you need some reminding of the general definition of ***t-statistic***, check out the definition on p. 404 of *AoS*). 

What assumption about the variances of the two groups are we making here?

A:We are comparing charges for people with insurance vs without insurance:

Let:
μ1 = mean charges for people with insurance
μ0= mean charges for people without insurance

Null hypothesis H0:
μ0=μ1
There is no difference in mean charges between insured and uninsured.

Alternative hypothesis HA:
μ0≠μ1
There is a difference in mean charges between insured and uninsured.
By using the pooled standard deviation, we are assuming:

σ0^2=σ1^2
The population variances are equal between the two groups.

__Q5:__ Perform this hypothesis test both manually, using the above formulae, and then using the appropriate function from [scipy.stats](https://docs.scipy.org/doc/scipy/reference/stats.html#statistical-tests) (hint, you're looking for a function to perform a _t_-test on two independent samples). For the manual approach, calculate the value of the test statistic and then its probability (the p-value). Verify you get the same results from both.

__A:__ 

```python
#Perform hypothesis test manually using formulae
from scipy import stats
# Split the charges by insurance claim
charges0 = medical.loc[medical['insuranceclaim'] == 0, 'charges']
charges1 = medical.loc[medical['insuranceclaim'] == 1, 'charges']

# Sample sizes
n0 = len(charges0)
n1 = len(charges1)

# Sample means
mean0 = np.mean(charges0)
mean1 = np.mean(charges1)

# Sample standard deviations
s0 = np.std(charges0, ddof=1)  # ddof=1 for sample SD
s1 = np.std(charges1, ddof=1)

# Pooled standard deviation
sp = np.sqrt(((n0 - 1)*s0**2 + (n1 - 1)*s1**2) / (n0 + n1 - 2))

# t-statistic
t_stat = (mean0 - mean1) / (sp * np.sqrt(1/n0 + 1/n1))

# Degrees of freedom
df = n0 + n1 - 2

# Two-tailed p-value
p_value = 2 * stats.t.sf(np.abs(t_stat), df)

print("Manual approach:")
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
#Perform hypothesis test by using function from scipy.stats
t_stat_scipy, p_value_scipy = stats.ttest_ind(charges0, charges1, equal_var=True)

print("\nSciPy t-test:")
print("t-statistic:", t_stat_scipy)
print("p-value:", p_value_scipy)
```

p-value ≥ 0.05 → fail to reject H0, no evidence of difference.There is no difference in mean charges between insured and uninsured. So we can interpret people with insurance really are charged same amount to those without.
